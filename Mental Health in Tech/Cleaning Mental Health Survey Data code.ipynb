```python
import os
import pandas as pd
```

```python
# import survey data

survey = pd.read_csv('mental_health_survey.csv')
survey.head()
```

```python
# survey metadata

survey.info()
```

```python
survey = survey.drop_duplicates()
survey.isna().sum() == 0
```

```python
survey_clean = survey[survey['Country'] == 'United States']
survey_clean = survey.drop(survey[survey['tech_company'] == 'No'].index)
```

```python
# checking for human errors when inputting age (only integar data type)

min_age = 18
max_age = 100

age_errors = survey_clean[(survey_clean['Age'] < min_age) | (survey_clean['Age'] > max_age)]
age_errors.head()
```

```python
# Get the indices of rows with age errors
age_error_indices = age_errors.index

# Remove the rows from the original DataFrame
survey_clean = survey_clean.drop(age_error_indices)
```

```python
survey_clean.to_csv("mentalhealth_clean.csv", index=False)
```
