```python
import pandas as pd
```

```python
# import survey data

survey = pd.read_csv('mental_health_survey.csv')
```

```python
# survey metadata

survey.info()
```

```python
# Checking for duplicate data

assert survey.duplicated().sum() == 0
```

```python
survey_clean = survey[survey['Country'] == 'United States']
survey_clean = survey_clean.drop(survey_clean[survey_clean['tech_company'] == 'No'].index)
survey_clean
```

```python
# Checking for human errors when inputting age (only integar data type)

# First converting our data type to 'string' so that we can use len() to count the characters. Then we are filtering out the data to keep only the rows where the length of 'Age' as a string is not equal to 2 characters
integer_outliers = survey_clean[survey_clean['Age'].astype(str).str.len() != 2]
integer_outliers

```

```python
# Checking to see if there are anymore outliers
min_age = 18
max_age = 100

age_errors = survey_clean[(survey_clean['Age'] < min_age) | (survey_clean['Age'] > max_age)]
age_errors.head()
```

```python
# Get the indices of rows with age errors
survey_clean = survey_clean.drop(age_errors.index)
survey_clean = survey_clean.reset_index(drop=True)
```

```python
survey_clean.to_csv("mentalhealth_clean.csv", index=False)
```
